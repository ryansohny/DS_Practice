## MNIST dataset

```python
from sklearn.datasets import load_digits # MNIST dataset load
digits = load_digits()  # 데이터 로드
digits.keys()
np.shape(digits['data']) # (1797, 64) # or digits.data

samples = [0, 10, 20, 30, 1, 11, 21, 31]
d = list()

for i in range(len(samples)):
    d.append(digits.images[samples[i]])

fig, axes = plt.subplots(ncols=8, figsize=(8,2), gridspec_kw={"wspace": 0.1}, constrained_layout=True)
for i in range(len(samples)):
    axes[i].imshow(d[i], interpolation='nearest', cmap='bone_r')
    axes[i].grid(False)
    #axes[i].tick_params(axis="both", length=0)
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    axes[i].set_title(f"image {i+1}")

fig.suptitle("Image of the number between 0 and 1")
```

```python
v = list()
for i in range(len(samples)):
    v.append(d[i].reshape(64,1)) # Vectorization

fig, axes = plt.subplots(1, 8, figsize=(8,3), gridspec_kw={"wspace": 0.1}, constrained_layout=True)
for i in range(len(samples)):
    axes[i].imshow(v[i], aspect=0.4, interpolation='nearest', cmap='bone_r')
    axes[i].grid(False)
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    axes[i].set_title(f"Vector {i+1}")

fig.suptitle("Vectorized Images", y=1.05) # y = 1.05 => Super title을 살짝 위로 올려줘서 axes들의 title과 겹치는 것 막아줌

```

## Tensors

```python
from scipy import misc  # Import misc package

img_rgb = misc.face()  # Upload color image
img_rgb
img_rgb.shape  # shape of the img_rgb (or np.shape(img_rgb))
```

```python
fig, axes = plt.subplots(2, 2, constrained_layout=True)
axes[0][0].imshow(img_rgb, cmap='gray')
axes[0][0].axis('off')
axes[0][0].set_title("RGB Color Image")

axes[0][1].imshow(img_rgb[:, :, 0], cmap='gray')
axes[0][1].axis('off')
axes[0][1].set_title("Red Channel")

axes[1][0].imshow(img_rgb[:, :, 1], cmap='gray')
axes[1][0].axis('off')
axes[1][0].set_title("Green Channel")

axes[1][1].imshow(img_rgb[:, :, 2], cmap='gray')
axes[1][1].axis('off')
axes[1][1].set_title("Blue Channel")
```
